## Trabalhando com o Geopandas

O Geopandas é um projeto de código aberto escrito em Python, possui bastante maturidade, eficácia e performance para trabalhar com grandes quantidade de dados. Ele usa o Pandas, NumPy e Shapely para trabalhar com dados georeferenciados e portanto achamos oportuno usa-lo.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

plt.rcParams['figure.figsize'] = (40, 30)

df_qf = gpd.read_file(f'gis/SIRGAS_SHP_quadraMDSF.shp')
df_qv = gpd.read_file(f'gis/SIRGAS_SHP_quadraviariaed_polygon.shp')
df_l = gpd.read_file(f'gis/SIRGAS_SHP_logradouronbl.shp')
df_s = gpd.read_file(f'gis/SIRGAS_SHP_setorfiscal.shp')
```

```python
df_s = df_s[df_s.st_codigo == '037']
```

```python
df_l = gpd.clip(df_l, df_s)
```

```python
df_vias = gpd.overlay(df_s, df_qv, how='difference')
```

```python
f, ax = plt.subplots(1)
df_vias.plot(ax=ax, alpha=0.5)
df_l.plot(ax=ax)
# df_logradouros.plot(ax=ax)

plt.show()

```

Obtendo os pontos finais e iniciais de cada logradouro podemos saber onde as ruas começam e terminam

```python
gdf_cruzamentos_inicio = df_l['geometry'].map(lambda x: (x.interpolate(0, normalized=True)))
gdf_cruzamentos_final = df_l['geometry'].map(lambda x: (x.interpolate(1, normalized=True)))
gdf_cruzamentos =  pd.concat([gdf_cruzamentos_inicio, gdf_cruzamentos_final])
gdf_c = gdf_cruzamentos
```

Como alguns pontos acabam ficando muito próximos, seria interessante agregar todos os pontos do cruzamento.

```python
cruzamentos = [p.centroid for p in gdf_c.buffer(5).unary_union]
# ds_cruzamentos = gpd.GeoSeries(cruzamentos)
# ds_cruzamentos.crs = {'init': 'epsg:31983'}
df_cruzamentos = gpd.GeoDataFrame(geometry=cruzamentos)
df_cruzamentos
```

```python
f, ax = plt.subplots(1)
df_vias.plot(ax=ax, alpha=0.4)
df_l.plot(ax=ax)
# df_cruz_buffer.plot(ax=ax, alpha=0.3, color='red')
df_cruzamentos.plot(ax=ax, lw=7, color='black')
df_cruzamentos.buffer(10).plot(ax=ax, alpha=0.3, color='red')
# Aplicando o zoom
plt.margins(x=-0.10, y=-0.40)

plt.show()
```

```python
df_pts = gpd.GeoDataFrame(geometry=df_cruzamentos.buffer(10).boundary.intersection(df_l.unary_union))
df_pts[df_pts.geometry.type == 'MultiPoint'].plot()
```

```python
df_p_center_x = df_cruzamentos[df_pts.geometry.type == 'MultiPoint'].geometry.apply(lambda x: x.coords[:][0][0])
df_p_center_y = df_cruzamentos[df_pts.geometry.type == 'MultiPoint'].geometry.apply(lambda x: x.coords[:][0][1])
```

```python
df_p_tang_x = df_pt_primeiro.geometry.apply(lambda x: x.coords[:][0][0])
df_p_tang_y = df_pt_primeiro.geometry.apply(lambda x: x.coords[:][0][1])
```

```python
import numpy as np
angle = np.degrees(np.arctan2(df_p_tang_y - df_p_center_y, df_p_tang_x - df_p_center_x))
angle
```

```python
# df_pts = df_pts[df_pts.geometry.type == 'MultiPoint']
# Rotacionando para o ponto inicial coincidir com um ponto de intersecção

from shapely.affinity import rotate, scale

df_bounds['angle'] = angle
df_bounds
df_bounds_rotated = gpd.GeoDataFrame(geometry=list(map(lambda x: rotate(x.geometry, x.angle, origin='center'), df_bounds.itertuples())))
df_bounds_rotated_p = gpd.GeoDataFrame(geometry=list(map(lambda x: Point(x.coords[:][0]), df_bounds_rotated.geometry)))
```

```python
f, ax = plt.subplots(1)
df_vias.plot(ax=ax, alpha=0.4)
df_l.plot(ax=ax)
# df_cruz_buffer.plot(ax=ax, alpha=0.3, color='red')
df_cruzamentos.plot(ax=ax, lw=7, color='black')
df_cruzamentos.buffer(10).plot(ax=ax, alpha=0.3, color='red')
df_bounds_rotated_p.plot(ax=ax, lw=7, color='green')
# Aplicando o zoom
plt.margins(x=-0.10, y=-0.40)
```

```python
df_bounds = gpd.GeoDataFrame(geometry=df_cruzamentos.geometry.buffer(10).boundary)

df_cruzamentos['id'] = df_cruzamentos.index.values
df_pts['id'] = df_bounds_rotated.index.values

df_pt_pts = df_cruzamentos.merge(df_pts, on='id', suffixes=('_cruzamento', '_pt'))

def draw_lines(item):
    if item.geometry_pt.type == 'MultiPoint':     
        return scale(
            MultiLineString(
            list(map(
            lambda x: 
            LineString((item.geometry_cruzamento.coords[:][0], list(x.coords)[0])), 
            item.geometry_pt.geoms))),
            xfact = 2.0,
            yfact = 2.0,
            origin = (item.geometry_cruzamento.coords[:][0]))
    elif len(item.geometry_pt.coords[:]) > 0:
        return scale(
            LineString((item.geometry_cruzamento.coords[:][0], item.geometry_pt.coords[:][0])),
            xfact = 2.0,
            yfact = 2.0,
            origin = (item.geometry_cruzamento.coords[:][0]))

l = list(map(lambda x: draw_lines(x), df_pt_pts.itertuples()))

df_cut_lines = gpd.GeoDataFrame(geometry=l)
```

```python
f, ax = plt.subplots(1)
df_vias.plot(ax=ax, alpha=0.4)
df_l.plot(ax=ax)
# df_cruz_buffer.plot(ax=ax, alpha=0.3, color='red')
df_cruzamentos.plot(ax=ax, lw=7, color='black')
df_cruzamentos.buffer(10).plot(ax=ax, alpha=0.3, color='red')
df_bounds_rotated_p.plot(ax=ax, lw=7, color='green')
df_cut_lines.plot(ax=ax, lw=7, color='orange')
# Aplicando o zoom
plt.margins(x=-0.10, y=-0.40)
```

## Criando linhas para 'cortar' os polígonos

Agora que temos que cortar o polígono precisamos de linhas para realizar esse fatiamento. 
Primeiramente vamos criar uma intersecção dos buffers dos nós com o traçado dos logradouros

```python
df_bounds['id'] = df_bounds.index.values
df_cut_lines['id'] = df_cut_lines.index.values
```

```python
from shapely.ops import split

df_arcs = df_bounds.merge(df_cut_lines, on='id', suffixes=('_bound', '_line'))
df_arcs_cutted = gpd.GeoDataFrame(geometry=list(map(lambda x: split(x.geometry_bound, x.geometry_line), df_arcs[df_pts.geometry.type == 'MultiPoint'].itertuples())))
# df_arcs_cutted.plot()

arcs_filtered = df_arcs_cutted.geometry.apply(lambda x: MultiLineString(list(filter(lambda y: y.length > 0.5, list(x.geoms)))))

df_arcs_filtered = gpd.GeoDataFrame(geometry=arcs_filtered)

p = df_arcs_filtered.geometry.apply(lambda x: MultiPoint(list(map(lambda y: y.interpolate(0.5, normalized=True), list(x.geoms)))))

df_mpoints = gpd.GeoDataFrame(geometry=p)
```

```python
f, ax = plt.subplots(1)
df_vias.plot(ax=ax, alpha=0.4)
df_l.plot(ax=ax)
# df_cruz_buffer.plot(ax=ax, alpha=0.3, color='red')
df_cruzamentos.plot(ax=ax, lw=7, color='black')
df_cruzamentos.buffer(10).plot(ax=ax, alpha=0.3, color='red')
df_bounds_rotated_p.plot(ax=ax, lw=7, color='green')
df_mpoints.plot(ax=ax, lw=7, color='orange')
# Aplicando o zoom
plt.margins(x=-0.30, y=-0.40)
```
